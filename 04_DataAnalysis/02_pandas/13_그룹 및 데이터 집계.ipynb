## aggregation : 집계
집계는 여러 데이터를 하나의 요약된 결과로 결합하고 통계를 내는 과정.

```python
# 여러 컬룸에 대해 서로 다른 집계함수를 수행
df.agg(
    나이평균=('나이','mean'),
    몸무게중위=('몸무게','median'),
    급여표준편차=('급여','std')
).round(2)
```

```python
# 시리즈를 만들면서 각각의 항목을 계산
result= pd.Series({
    '나이평균':df['나이'].mean(),
    '몸무게중위':df['몸무게'].median(),
    '급여표준편차':df['급여'].std(),
}).round(2)
result
```

```python
# 집계함수 중 파라미터(인자)가 있어야만 하는 함수 quantile 을 사용하려면 문자열만으로는 불가능하고 람다함수를 만들면 됨
df['급여'].agg([
    'mean',
    'median',
    lambda x:x.quantile(0.25),
    lambda x:x.quantile(0.50),
    lambda x:x.quantile(0.75),
])
```

```python
df['급여'].agg(
    mean='mean',
    median='median',
    q1=lambda x:x.quantile(0.25),
    q2=lambda x:x.quantile(0.50),
    q3=lambda x:x.quantile(0.75),
)
```

```python
# groupby() + egg()
# 부서별 지출 데이터는 컬룸마다 의미가 다르기에 컬럼별로 다른 집계함수를 쓰는게 자연스러움.

# 부서별 지출 데이터 다시 불러오기
df= pd.read_csv('./data_set/expense_data.csv')
df
```

```python
# 컬룸마다 집계해야할 방법이 다를 수 있음.
# 금액은 '규모'
# 수량은 '빈도'
# 제품(항목)은 '다양성'

# 부서별 지출 요약 보고 레포트...
dept_summary= df.groupby('department').agg(
    총지출액= ('expense_amount','sum'),    #전체 비용 규모
    평균지출액= ('expense_amount','mean'), #지출 패턴
    최고지출액= ('expense_amount','max'),  #고액 지출 규모
    총구매수량= ('quantity','sum'),        #구매 빈도
    지출항목= ('product','nunique'),       #취급하는 항목 다양성
    담당직원수= ('employee','nunique'),    #직원수 규모 확인.. 관리 복잡도..
).round(1)

dept_summary
```

```python
# 건별 평균 지출내역 계산해보기.. 새로운 컬룸에 계산결과를 할당
df.groupby('department').agg(
    총지출액=('expense_amount','sum'),
    총구매수량=('quantity','sum'),
).assign(
    건당평균지출= lambda x:x['총지출액']/x['총구매수량']
)
```

### transform
그룹별로 계산은 하되, 결과를 '원래 행 개수 그대로'유지해줌.   

- 핵심차이   
agg() : 계산 후, 행 수 줄어듦   
transform() : 행 수 그대로 유지.

```python
# agg()
df.groupby('department')['expense_amount'].mean().round()   #밑에꺼랑 같음
df.groupby('department')['expense_amount'].agg('mean').round()
#부서별로 1행
#원본 데이터셋과 함께 비교가 안됨
```

```python
# transform()  --  각 행에 '부서 평균지출' 붙이기
df['부서평균지출']= df.groupby('department')['expense_amount'].transform('mean').round()
df
```

```python
# 개별지출이 부서의 평균지출보다 큰가? -- 의심.. 사기..
df.assign(평균초과여부= lambda x:x['expense_amount']>x['부서평균지출'])
```

```python
# 이상지출 탐지 ( z-score : 표준점수 - 너의 편차가 표준편차에 비해 얼마나 차이가 있는지 [(원점수-평균)/표준편차] )
df.assign(
    z_score= lambda x:
        (x['expense_amount']-x.groupby('department')['expense_amount'].transform('mean'))/x.groupby('department')['expense_amount'].transform('std')
)
```

```python
# z-score 해석
#  0 : 부서 평균과 같은
#  1 : 평균보다 1표준편차 큰 지출
# -1 : 평균보다 1표준편차 적은 지출

# z-score는 금액의 크기가 아니라 부서 지출 기준으로 볼때 상대적인 사용 정도를 보여주는 값!
```

```python
# 표준편차만으로는 올바른 분산의 정도를 파악하지 못할 수 있음.
# --> 변동계수 CV = 표준편차 / 평균 * 100
# 두 종류의 자료값이 차이가 있을때 단순하게 표준편차를 통해 분산정도를 파악할 수는 없음.

# 예) 성인여성 50명, 10세 소녀 50명을 대상으로 몸무게를 조사한 결과..
# 성인여성: 평균[51.1kg], 표준편차[5.0kg]
# 10세 소녀: 평균[26.49kg], 표준편차[3.66kg]

# 표준편차만 보면 성인여성 표본집단의 몸무게 변동이 더 커보임..
df= pd.DataFrame({
    '평균':[51.1, 26.49],
    '표준편차':[5.0, 3.66]
}, index=['성인여성','10세소녀'])
df
```

```python
# 변동계수 계산... --파생지표 추가
df['변동계수']=(df['표준편차']/df['평균'])*100
df
```

```python
# 예) 여행지를 추천하는 서비스 개발을 위해 성별이나 나이에 따라 여행지 선호를 분석하기!!

# 성별과 나이에 따른 여행지 선호 설문데이터
df= pd.read_csv('./data_set/travel_preference_survey.csv')
df
```

```python
# 성별 별로 3명씩 랜덤 샘플 추출
gender_sample= df.groupby('성별').sample(n=3, random_state=42)
gender_sample
```
