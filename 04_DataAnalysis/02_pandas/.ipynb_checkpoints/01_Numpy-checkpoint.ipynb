### 1. 일반적인 파이썬 배열(리스트)

```python
aaa=[10,20,30]
bbb=[4,5,6]

print(aaa)
print(bbb)
```

```python
#파이썬 리스트의 덧셈을 하면.. 수학계산이 아니라 결합연산
ccc= aaa+bbb
print(ccc)
```

```python
# 자료형 확인
print(type(aaa))
```

### 2. numpy 배열

```python
import numpy as np
aa= np.array([10,20,30])
bb= np.array([4,5,6])

print(aa)
print(bb)
```

```python
# 배열의 사칙연산이 수학적으로 수행됨
cc= aa+bb
print(cc)
```

```python
# 배열의 요소 개수 확인
print(len(aa))
print(len(bb))
print(len(cc))
```

```python
# 단, 넘파이 배열을 사용할때는 가급적 사이즈를 .shape 속성으로 확인하는 것을 권장
print(aa.shape)
print(bb.shape)
print(cc.shape)  # 튜플로 (행,열) 개수를 표시... 단, 현재 1차원이니.. 값이 하나만..
```

```python
#자료형 확인
print(type(aa))
```

### 3. 다차원 넘파이 배열

```python
aaaa= np.array([[10,20,30,40],[5,6,7,8]])
print(len(aaaa))  #2 <-- 행 갯수
print(aaaa.shape) #튜플 (행,열)

bbbb=np.array([[100,200,300,400],[50,60,70,80]])
print(bbbb.shape)

#사칙연산 가능
cccc= aaaa+bbbb
print(cccc.shape)
print(cccc)
```

```python
#사칙연산 모두 가능
print(aaaa-bbbb)
print(aaaa*bbbb)  #요소별 곱셈
print(aaaa/bbbb)  #요소별 나눗셈
```

```python
# 연산 주의! 행렬의 갯수가 다르면 에러!!!
dddd= np.array([10,20,30,40],[500,600])
```

```python
#연산하는 행렬의 갯수도 맞아야 함
eeee= np.array([[10,20],[50,60]])
ffff=aaaa+eeee
```

```python
# 수학적 행렬의 곱셈 - 1차원 배열일 때는 내적곱(dot), 2차원 배열일때는 수학의 행렬곱셈을 수행 
#2차원 배열
A= np.array([
    [1,2],
    [3,4]
])
B= np.array([
    [5,6],
    [7,8]
])
print( A @ B)  # 행렬곱
# or
print(np.matmul(A,B))
```

```python
# 1차원 배열일 경우에는 내적곱(요소끼리 곱하여 덧셈) -- 벡터 내적 --> 스칼라값
C=np.array([1,2,3,4])  #벡터
D=np.array([5,6,7,8])  #벡터
print(np.dot(C,D))     #스칼라
```

```python
#수학 및 통계 연산 기능 메소드 (함수)

# 10명의 학생 성적 데이터
scores= np.array([70,80,64,73,42,98,27,40,85,100])
```

```python
#요소들의 총합 구하기 [2가지 방법]
print('총점:',scores.sum())
print('총점:',np.sum(scores))
```

```python
# 요소들의 평균 구하기 [2가지 방법]
print('평균:',scores.mean())
print('평균:',np.mean(scores))
```

```python
# 요소들의 최대값/최소값 구하기 [2가지 방법]
print('최대값:',scores.max())
print('최대값:',np.max(scores))
print('최소값:',scores.min())
print('최소값:',np.min(scores))
```

```python
# 조건에 따라 원소들의 결과값을 계산하는 기능(파이썬 삼항연산자+반복문)
#예) 60점 이상은 'pass' 아니면 'fail'
result= np.where( scores > 60, 'pass', 'fail' )  # 조건, 참일때 값, 거짓일때 값
print(result)
```

```python
#배열 결함 (파이썬의 리스트처럼)
aaa= np.array([1,2,3])
bbb= np.array([4,5,6])

result= np.concatenate((aaa,bbb))  # 튜플로
print(result)
```

```python
# 2차원 배열일때는.. 행으로 붙일지.. 열로 붙일지..
a= np.array([[1,2],[3,4]])
b= np.array([[5,6],[7,8]])
```

```python
# 행 기준 연결(위아래)
result1= np.concatenate((a,b), axis=0) # axis=0 행방향으로 내려가면서 붙이기
print(result1)
```

```python
# 열 기준 연결(좌우)
result2= np.concatenate((a,b), axis=1) # axis=1 열방향으로 오른쪽으로 가면서 붙이기
print(result2)
```

```python
# 인덱싱 및 슬라이싱

#1번 학생의 성적데이터
first_student= scores[0]

#3~7번 학생(슬라이싱)  [인덱스 번호로는 2~6]
middle_students= scores[2:7]

print(first_student)
print(middle_students)
```

```python
# 불리언 인덱싱 - 기준점수 이상인 학생만 추출
# 80점 이상인 학생들의 성적만 추출
high_scores= scores[ scores>=80 ]  # 조건값이 true인 요소들만 인덱싱
print(high_scores)
print(high_scores.mean())
print(len(high_scores))
```

```python
# 배열의 차원을 변경 시켜주는 기능 reshape()
# 1차원 → 2차원 (reshape)
# 5명 학생 × 7일 = 35개 점수
scores_1d = np.array([
    80, 75, 90, 85, 87, 92, 78,
    70, 65, 75, 80, 82, 85, 79,
    90, 88, 91, 94, 96, 95, 93,
    60, 65, 70, 72, 74, 76, 78,
    85, 87, 88, 90, 92, 91, 89
])
print(scores_1d)
```

```python
#(학생수, 날짜수) 형태로 변경
scores_2d= scores_1d.reshape(5,7)  #5행 7열로 만들어줘.
print(scores_2d)
```

```python
# 행, 열 중 한쪽만 지정하고.. 나머지는 알아서 계산되도록 설정
scores_2d= scores_1d.reshape(5,-1)  #5행 ?열로 만들어줘.
print(scores_2d)
```

```python
scores_2d= scores_1d.reshape(4,-1)  #4행 ?열로 만들어줘.
print(scores_2d)
```

```python
scores_2d= scores_1d.reshape(-1,7)  # ?행 7열로 만들어줘.
print(scores_2d)
```

```python
# 차원을 변환하면 좋은점.. 분석하기 용이해짐

#학생별 평균점수
student_aver= scores_2d.mean(axis=1)  # 행 기준 계산 -- 옆으로 가면서 계산
print(student_aver, len(student_aver))

# 날짜별 평균 점수
day_aver= scores_2d.mean(axis=0)  # 열 기준 계산 -- 아래로 가면서 계산
print(day_aver, len(day_aver))
```
